```python
# 예제 이용해 for 이해하기
test_list = ['one', 'two', 'three']
for i in test_list:
    print(i)
    
```

```python
for 숫자 in range(1,3,1):
    print(숫자, end=" ")
```

```python
# for 딕셔너리 처리
# 딕셔너리 자체는 인덱스가 없다.
# 단, 딕셔너리 key / value / keys를 뽑아내면 
# 이터러블한 인덱스가 생긴다.
점수 = {'국어': 90, '영어': 85, '수학': 95}
# for aaa in 점수.keys():
for 과목 in 점수:
    print(과목, end=" ") # key값이 출력된다.   
```
