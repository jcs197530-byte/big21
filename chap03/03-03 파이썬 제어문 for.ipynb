```python
# for 딕셔너리 처리
# 딕셔너리 자체는 인덱스가 없다.
# 단, 딕셔너리 key / value / keys를 뽑아내면 
# 이터러블한 인덱스가 생긴다.
점수 = {'국어': 90, '영어': 85, '수학': 95}
# for aaa in 점수.keys():
# 키값으로 반복
for 과목 in 점수:
    print(과목) # key값이 출력된다.
# 값으로 반복
for 점수값 in 점수.values():    
    print(점수값) # key값이 출력된다.

# 제일 많이 사용 : items()=>[(과목,점수값),(과목,점수값),...]
for 과목,점수값 in 점수.items():
    print(f'{과목} : {점수값} 점')

    
```

```python
# set도 for 사용할 수 있다
# ()로 값이 들어가면 set
# ()에 key,value가 들어가면 딕셔너리
# set은 인덱스가 없기 때문에 출력되는 순서는 보장되지 않는다
# set은 데이터확인정보(중복데이터 존재하지 않는다)
태그 = set(['python', 'ai', 'for'])
for 태그값  in 태그:
    print(태그값)
```

```python
# for 문에 break, continue, else 
# 1. break : 반복문을 중단한다.
# 1부터 5까지 출력, 3이면 중지
for 숫자 in range(1,6):
    if 숫자 == 3:
        break       
    print(숫자)

결과_리스트 = [숫자 for 숫자 in range(1, 6) if 숫자 < 3]
print(결과_리스트)

```

```python
# continue : 반복 건너뜨기
# 1부터 5까지 출력, 3이면 건너띄기 => 시작 : 1 , 끝 : 5+1
for 숫자 in range(1,6):
    if 숫자 == 3:#숫자가 3이면
        continue
    print(숫자)
```

```python
# else : 
# 리스트 안에 문제가 있는 값이 있다. 
# 문제가 있는 값이 나오면 반복 중지 , 그 값이 무엇인지 출력해라

# 리스트 안에 홀수값이 나오면 중지
# else : 블록은 break가 안 되었을때 실행할 코드 넣는곳
# 문제가 없는 데이터라고 인식을 하려고 하는 의미라고 보면 된다.
# 이 부분은 정말 잘해야 한다.
# else는 내가 원하는 데이터 이외의 데이터가 들어왔는지 검증하는 함수라고 보면된다.
# 예 : 짝수만 들어와야 되는 데이터인데, 
# 짝수 이외의 데이터가 들어왔는지 체크하는 함수라고 보면된다.
숫자들 = [4,6,2,8]

for n in 숫자들:
    if n % 2 == 1: # 홀수이면(나머지 연산자 이용)  
        break      
    print(n)
else:
    print("홀수가 없습니다")
```

```python
# 리스트 컴프리헨션(필수 기능)
# 간결하게 사용하기 위함 / 1줄로 끝날수 있다
# for 문으로 새 리스트를 만드는 작업을 한줄로 압축하는 문법
# 리스트 컴프리헨션의 목적은 새 리스트를 만드는 것이다.
# 기본 예제 : 어떻게 작동하는지 확인
# [1,2,3,4] 존재, 이 리스트를 가지고 [2,3,6,8] 새 리스트를 생성
# 1. 사용 안하고
# 숫자들 = []
# for n in range(1, 5):
#     숫자들.append(n*2)
# print(숫자들)
data = [1,2,3,4]
new_data = []
for num in data:
    new_data.append(num*2)
print(new_data)

# 2. 하고
# 숫자들 = [num*2 for num in data]
# print(숫자들)
new_data = [num * 2 for num in data]
print(new_data)
```
