```python
# 조건문 : if, switch, match ~ case(제약사항이 있다)
# 기본적으로 if 사용한다, 복잡해지면 match 고려한다
# 3가지 구조의 if가 있다
# 1. if 구조 : 특정 조건에 만족할 경우에만 어떤 코드 실행하고자 할때
#    if 조건식:
#     조건식이 True 일 때 실행할 코드 블록
#
# 2. if ~ else
#    if 조건식:
#     조건식이 True 일 때 실행할 코드 블록
#    else:
#     조건식이 False 일 때 실행할 코드 블록
# 3. if ~ elif ~ else
#    if 조건식1:
#     조건식1이 True 일 때 실행할 코드 블록
#    elif 조건식2:
#     조건식2가 True 일 때 실행할 코드 블록
#    else:
#     모든 조건이 False 일 때 실행할 코드 블록
```

```python
# '돈이 있으면 택시를 타고, 돈이 없다면 걸어간다'
# 변수명 선언은 : 영어, 카멜표기법 사용
# 돈 저장할 변수 필요 : money =0, 1 0 : 없다, 1 : 있다
money = 0 # 돈이 있다를 의미한다
if money: # ":" 기능 -> True 일 때 블록의 시작, 블록 : 여러 명령을 묶는 역할
    print('택시를 타고 가라')
else: # False 일때 블록의 시작, 
    print('걸러 가라')
```

```python
# "만약 5000원 이상의 돈을 가지고 있으면 택시를 타고 그렇지 않으면 걸어 가라"
money = 5000
if money >= 5000:
    print('택시를 타고 가라')
else:
    print('걸어 가라')
```

```python
# "돈이 5000원 이상 있거나 카드가 있으면 택시를 타고 그렇지 않으면 걸어 가라"
money = 4000 # 돈은 4000원을 가지고 있다
card = 1 # card 를 가지고 있다
if money >= 5000 or card: # 돈을 5000원 이상 가지고 있거나, 카드 있거나
    print('택시를 타고 가라')
else: # False 인 경우 : 돈이 5000원 안되고, 카드도 없을 경우
    print('걸어 가라')
```

```python
# in 연산자 => 특정값 in iterable(반복가능한 객체) 사용
# iterable : 여러개 저장하고 있는 놈, 반복 가능한 놈, 인덱스 가지고 있는 놈
# list, tuple, string <= iterable 한 놈들이다.
print(1 in [1,2,3]) # 1,2,3 이 들어있는 리스트에서 1 을 찾은 것 
print(5 in [1,2,3]) # 1,2,3 이 들어있는 리스트에서 5 을 찾은 것 
```

```python
# "주머니에 돈이 있으면 택시를 타고, 없으면 걸어 가라"
pocket = ['paper','cellphone','money']
if 'money' in pocket: # 주머니에 안에 돈이 있으면
    print('택시를 타고 가라')
else:
    print('걸어 가라')
```

```python
# 돈이 있거나 카드가 있으면 택시 타고, 둘 다 없으면 걸어 가라
# if money or card:
pocket = ['paper','cell']
if 'money' in pocket or 'card' in pocket:
    print('택시 타고 가라')
else:
    print('걸어 가라')
# 주머니에 돈이 있으면 택시를 타고
# if money: print('택시')
# 주머니에 돈은 없지만 카드가 있으면 택시를 타고
# else if card : print('택시')
# 돈도 없고 카드도 없으면 걸어 가라
# else: print('걸어 가라')
if 'money' in pocket:
    print('택시 타고 가라')
elif 'card' in pocket:
    print('택시 타고 가라')
else:
    print('걸어 가라')
```
