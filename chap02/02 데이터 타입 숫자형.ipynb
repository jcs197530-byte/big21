```python
# : 한 줄 주석 표시할 때 사용합니다.
# 파이썬 자료형 숫자형 : 
# 숫자형 자료형
# 자료형 : Data Type , 저장 공간을 생성하라는 명령 , 이름 부여한다.
# 숫자 자료형 : 정수(int) , 실수(소수점 존재,float) 나누어진다
# print(type(10))
# print(type(10.))
# 10 / = / int_num 순서대로 진행됨.
# 동적활동 / java는 정적활동
# 할당 타입 찾고 / 어떤 형태로 대입할지 / 공간의 이름(변수명)
# 공간이 없으면 대입이 되지 않는다.
# 파이썬은 대입될 때 변수명이 생성된다. --> 동적할당
# num = 10 # 이 때 num 은 정수 공간에 부여된 변수명
# num = 10. # 이때 num은 실수 공간에 보여된 변수명(다른 변수값을 넣으면 변수 타입 자체가 변환된다. int -> float)
# num = '10' # 이때 num은 실수 공간에 보여된 변수명(다른 변수값을 넣으면 변수 타입 자체가 변환된다. int -> float)
# java 언어의 경우 : int num = 10 / int 1 / num 2 / 10 3
# 파이썬은 변수타입을 변경할 수 있다.(다른 언어는 정적할당)
# str 문자열 object
# 어드레스 주소 할당(주소값을 저장) : str '10'(200번주소) / int '10'(100번주소)
# num의 주소를 찍고 있음.
# 단어를 붙이는게 LLM이다.
# 단어를 붙여서 결과를 보여지게 된다.
num = 10 # 이 때 num 은 정수 공간에 부여된 변수명
# 10이라는 공간 만들 / 이름은 num / 이름에 
num = '10' # 이 때 num 은 정수 공간에 부여된 변수명

print(num + str(10333))
# print(type(num))

```

```python
# 정수 데이터 타입
# 정수 : 음의 정수 , 양의 정수 , 0
a = 123 # 123 양의 정수를 a에 대입 , 목적 a로 123 정수르 사용
a = -123 # 음의 정수 -123을  a의 주소값에 대입한 상태다, 
a = 0
```

```python
# 실수 데이터 타입
a = 1.2
a = -3.45
a = 0.0

a = 4240000000000. #지수
a = 4.24E10 #E10 .4.24다음에 10자리뒤로 . 이동해라

a = 42400000000000000000000. #지수
a = 4.24E10 #E20 .4.24다음에 20자리뒤로 . 이동해라

# 지수값 처리가 많기에 반드시 알아야 할 내용이다.
a = 4.24E2 #4.24 -> 424.
a = 4.24E-2 #4.24 -> 0.0424

```

```python
# 16진수 : 0 - 15(F)의 값으로 숫자를 표기하는 방법
# 한자리 값으료 표기 : 0-9까지는 한 자리로 표기할 수 있다. 
# 10 -> A ,  11 -> B , 12 -> C , 13 -> D , 14 -> E , 15 -> F
# 표기법 : 0x로 시작해야 한다
a = 0xAB #0x로 시작한다는 뜻은 A(10)하고 B(11)가 16진수 값이다라는 의미다.
#2진수 표기했을때 길어진다. -> 16진수로 처리하면 줄어든다. 유니코드에서 16진수 사용한다.
a = 171 #0xAB(16진수) -> 171(10진수) #16진수를 10진수로 바꾸는 방법

```

```python
# 연산자
# 사칙연산(+,-,*,/)
a = 3 # a에 3을 넣어 놓았다
b = 4
print(a + 4)
print(a / b) #정수/정수 -> 실수가 나온다
```

```python
# 거듭제곱  연산자 : **
a = 2
b = 4
print(a**b)
```

```python
# 나머지 구하는 연산자 : % 
# 배수 처리할때 많이 사용함
# 0 이 나오는값이 배수이다.
# 짝수와 훌수 구분하려고 할때 많이 사용함
a = 7
b = 3
print(7 % 3)


```

```python
# 나눗셈 후 소수점 아랫 자리를 버리는 연산자 : (단, 양수/음수의 작은값으로 반환)
# 나누기
a = 7
b = 4
print(a/b)
print(a//b) # // 나누기 한 후에 소숫점 자리는 버리기 1.75 -> 1(단, 작은값으로 반환)

a = -7
print(a/b)
print(a//b) # // 나누기 한 후에 소숫점 자리는 버리기 -1.75 -> -2(단, 작은값으로 반환)
```
