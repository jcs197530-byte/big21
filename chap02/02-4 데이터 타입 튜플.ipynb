```python
# 튜플 vs 리스트 : 인덱스 존재 => 인덱싱,슬라이싱 가능
# 튜플은 불변 / 리스트는 가변
# 튜플 생성
t1 = () # 빈 소괄호 => 빈 튜플 오브젝트 생성된다
t2 = (1,) # 1이 들어있는 튜플 오브젝트 생성된다
t3 = (1,2,3) # t3 = 1,2,3 동일 
# 1,2,3은 튜플이다. 불변이다. 이 부분을 반드시 기억해 주세요.
# print(t2)
```

```python
# 불변 확인 : 삭제 안되는거 확인
t1 = (1,2,3,4) #소괄호는 빈 튜플 오브젝트 생성된다
t1 = 1,2,3,4 #괄호가 아예없우면 기본이 튜플 오브젝트 생성된다
# del t1[0]
# t1[0] = 5
```
