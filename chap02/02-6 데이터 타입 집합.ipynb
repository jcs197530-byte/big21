```python
# set([]) 함수 사용 , 여러개 저장
s1 = set([1,1,2,2,3,4]) # 집합 오브젝트 생성, 단, 중복 제거하고 생성
print(sorted(s1)) #
print(list(s1)[0])
print(list(s1)[1])
print(list(s1)[2])
print(list(s1)[3])
# 1,000개의 물품이 판매되었는지 알고 싶다.

s2 = set('Hello')
print(s2) 
```

```python
# 집합(set) : 중복 제거 , 합집합 , 교집합, 차집합
# 교집합 - 합집합 순으로 알고리즘이 추천(집합연산자) 
# 교집합이 제일 중요함. 알고리즘 1순위
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1 & s2) # 교집합
print(s1.intersection(s2)) # 교집합

print(s1 | s2) # 합집합
print(s1.union(s2)) # 합집합

print(s1 - s2) # 차집합(독특한 특성)
print(s1.difference(s2)) # 차집합(독특한 특성)

print(s2 - s1) # 차집합(독특한 특성)
print(s2.difference(s1)) # 차집합(독특한 특성)
```

```python
# 집합 메소드 
# add(추가) , update(수정) , remove9(제거)
# update : 여러개 추가의 의미. (의미가 약간 다름) 
# update -> 다른 언어의 append 같은 개념으로 봐야한다.
# 그룹으로 추가한다는 의미로 인지해야 한다.
s1 = set([1,2,3])
s1.update([4,5,6])
print(s1) # {1,2,3} 중복 제거됨.
```

```python
s1 = set([1,2,3])
s1.update((4,))
print(s1)
# update(iterable : 반복 가능한 클래스(정수실시처럼 수정불가한건 안됨.) )
# iterable class :  반복 가능한 클래스
# 1.리스트는 iterable을 상속 받은 놈(아무거나 저장.수정가능)
# 2.투플은 리스트 상속 받은 놈(리스트 상속. 아무거나 저장.수정불가)
# 3.문자열은 튜플 상속 받은 놈(튜플 상속. 수정불가)
```
