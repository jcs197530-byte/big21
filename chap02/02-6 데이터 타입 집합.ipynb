```python
# set([]) 함수 사용 , 여러개 저장
s1 = set([1,1,2,2,3,4]) # 집합 오브젝트 생성, 단, 중복 제거하고 생성
print(sorted(s1)) #
print(list(s1)[0])
print(list(s1)[1])
print(list(s1)[2])
print(list(s1)[3])
# 1,000개의 물품이 판매되었는지 알고 싶다.

s2 = set('Hello')
print(s2) 
```

```python
# 집합(set) : 중복 제거 , 합집합 , 교집합, 차집합
# 교집합 - 합집합 순으로 알고리즘이 추천(집합연산자) 
# 교집합이 제일 중요함. 알고리즘 1순위
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1 & s2) # 교집합
print(s1 | s2) # 합집합
print(s1 - s2) # 차집합
```
