```python
# 리스트의 인덱싱과 슬라이싱
# 인덱싱 : 단일값 추출
# 슬라이싱 : 여러개값 추출, [처음:끝값-1]
a = [1,2,3]
print(a)
# 3 추출
print(a[2])
print(a[-1])
```

```python
# 
a = [1,2,3,['a','b','c']]
print(a[-1])
print(a[3][2])
print(a[1])
```

```python
# 슬라이싱
a = [1,2,3,4,5]
print(a[0:2])
print(a[:2])
print(a[2:5])
print(a[2:])
```

```python
# 어떤 반에 3명의 학생이 있다. 이름과 국어점수 저장 출력
students = [
    ["홍길동","고길동","마이꼴"], # 이름 저장하는 리스트
    [80,90,60] # 이름 저장하는 리스트
]

print(students)
print(students[0][1])
print(students[1][1])
print(students[0][0],":",students[1][0])
print(students[0][1],":",students[1][1])
print(students[0][2],":",students[1][2])
```

```python
a = [1,2,3,4,5]
# a[0] = 10
b = a
c = a[:] #전체를 복사해서 대입
print(id(a),id(b),id(c))
```

```python
# 리스트 연산자
a = [1,2,3]
b = [4,5,6]
# 두개 리스트 합치기(+) : 사용하지 마세요.
# 구현된 메소드를 사용하는 것이 좋다.
print(a+b)
# 리스트 반복 : *
print(a * 3)
```

```python
# 리스트의 수정, 변경 , 삭제
# 1. 하나의 값(단일값:인덱싱) 삭제
a = [1,2,3]
a[2] = 4
print(a)
```

```python
# 2. 리스트에서 연속된 범위의 값(슬라이싱) 수정
a = [1,2,4]
# print(a[1:2]) # 리스트 2 -> 슬라이스는 리스트로 나옴
# print(a[1])   # 단일값 2 -> 단일값은 정수가 나옴
a[1:2] = ["a","b","c"] #슬라이싱 개념으로 들어가면 그 부분을 모두 덮어씀
# a[1] = ["a","b","c"] #1개 문자열에 대입하면 리스트로 덮어씀
print(a)
```

```python
# 리스트 요소 삭제
a = [1, 'a', 'b', 'c', 4]
a[1:3] = []
# a.remove('a')
print(a)
```

```python
# 리스트 요소 삭제 : del 함수 사용
# del 함수는 모든 오브젝트를 삭제할 수 있다.
a = [1, 'a', 'b', 'c', 4]
b = "aabbcc"
del a[1:3]
# del b("aa")
print(a)
# print(b)
```

```python
# 리스트 관련 메소드들
# 1. 리스트 마지막에 추가 : append(추가할 값)
a = [1,2,3]
a.append(4)
print(a)
# 현재 리스트 안에 리스트 추가 [1, 2, 3, 4,[5,6]]
a.append([5,6])
print(a)
```

```python
# 리스트 정렬 : 
# sort() -> 오른차순 정렬만 [1, 2, 3, 4]
# 내림차순 [4, 3, 2, 1]
# 슬라이싱을 
a = [1,4,3,2]
print(id(a))
a.sort()
print(id(a))
print(a)
print(a[::-1])
print(id(a[::-1]))
print(a)
print(id(a))
```

```python
# 알파벳 순서로 정렬 : 오른차순만 가능하다
a = ["a","c","b"]
print(id(a))
a.sort()
print(a)
print(id(a))
print(a[::-1])
print(id(a[::-1]))
```

```python
# 리스트 뒤집기 : reverse(), 정렬은 아니다
a = ["a","c","b"]
a.reverse()
print(a)
a.sort(reverse=True)
print(a)
```

```python
# 위치 반환 : index(찾을값) => 
# 찾을 값이 있을경우 : 해당 위치의 인덱스 값을 반환 , 
# 없을경우 : 예외(오류) 발생
a = [1,2,3]
print(a.index(1))
# print(a.index(4))


```

```python
# 리스트에 요소 삽입 : insert(삽일할 위치 , 삽일할 값)
a = [1,2,3]
# 첫번째 자리에 4를 삽입
a.insert(0,4)
print(a)
a.insert(len(a)-1,5)
print(a)
```

```python
# 리스트 요소 제거 : remove(제거할 값)
a = [1,2,3,1,2,3]
# 3 제거
a.remove(3)
print(a)
a.remove(3)
print(a)
```

```python
# 리스트에서 특정값을 끄집어내기 : pop(), pop(특정 인덱스)
# pop(x) : x -> 값을 안 준 경우다 , 마지막 값을 끄집어낸다.
a = [1,2,3] # 제일 많이 사용
# a.pop(1) 제일 사용안함
print(a)
```

```python
# 리스트에 포함된 값의 개수 세기 : count(찾을값)
a = [1,2,3,1]
a.count(1) # 1이 2개다
```

```python
# 리스트 확장 : extend(리스트)
a = [1,2,3]
b = [4,5,6]
# print(a.extend(b)) #리턴값이 none
# print(a)
b.extend(a) #b에 a값을 extend
print(b)
print(a)

```
