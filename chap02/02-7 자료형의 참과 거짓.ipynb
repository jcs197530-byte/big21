```python
# 조건식 자리에 데이터 타입이 값이 들어가면 참/거짓의 역할을 한다.
# 데이터 타입(변수)에 값이 있으면 참, 없으면(비었으면) 거짓 역할을 한다

# 숫자
# 기본 if
# if 조건식 자리:
#   True 일때 실행될 코드가 들어갈 자리
# else:
#   조건식이 False 일때 실행될 코드가 들어갈 자리
num = 0
if num:
    print("num의 값이 0이 아니다.")
else:
    print("num의 값이 0이다.")
```

```python
list_= [1] # 빈 리스트
if list_:
    print("빈 리스트가 아니다.")
else:
    print("빈 리스트다.")
```

```python
a = [1, 2, 3,4]
# while 조건식자리 : 현재상태 조건식 자리에 리스트
# 조건식 Ture일때 아레 코드를 실행하는 것이 while 이다.
while a:  
    print(a.pop()) # 리스트의 마지막 요소를 꺼내서 반환 => 리스트가 비어있지 않으면 참이므로 반복문이 계속 실행된다.
```
