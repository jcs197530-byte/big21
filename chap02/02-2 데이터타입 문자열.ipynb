```python
# 고급 문자열 포매팅
# 포매팅 3가지 방법
# 1. % 사용하는 방법 --> 사용하면 안됨
# 2. format() 메소도를 사용하는 방식 --> 오늘 설명할 내용
# 3. f-string 사용 -> 다양하게 사용

# name = "민수"
# age = 45
# # 문자열 '앞'에 f를 붙이는 것이 핵심!
# result = f"이름은 {name}이고, 나이는 {age}살입니다.{name}{name}{name}"
# print(result) # 출력: 이름은 민수이고, 나이는 45살입니다.
```

```python
# format() 메소드 사용
num = 3
a = "I eat %d apples" % num
print(a)

b = "I eat {} apples{}{}"
print(b.format(num,num,num))

print(f"I eat {num} apples{num}{num}") # 처리 속도가 빠르다
```

```python
# f-string
# 정렬 : 왼쪽,오른쪽,가운데
h = "Hello"
print(h)
print(f"{h}")
print(f"{h:10}") #10자리 생성하고 찍어라 , 기본 왼쪽정렬
print(f"{h:<10}") #10자리 생성하고 찍어라 , 왼쪽정렬
print(f"{h:>10}") #10자리 생성하고 찍어라 , 오른쪽정렬
print(f"{h:^10}") #10자리 생성하고 찍어라 , 가운데정렬
print(len(f"{h:10}")) #10자리 생성하고 찍어라 10자리
```

```python
# f-string 에서 {} 표현식 이라고 한다. 중괄호안에 식(코드)가 들어갈 수 있다.
# 조건식 부여 : if 문
h = "hello"
b = "bye"

print(f"{h if len(h) > len(b) else b}") # 조건식이 true이면 앞내용 실행
print(f"{h if len(h) < len(b) else b}") # 조건식이 false이면 뒷내용 실행
```
