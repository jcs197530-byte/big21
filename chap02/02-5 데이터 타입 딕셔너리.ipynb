```python
# 딕셔너리 : 데이터 타입 -> 저장소 : 변수(속성 attribute)
# 데이터 타입별로 저장하는 방식과 종류 , 형태가 상이하다
# 숫자형 : 정수 , 실수 저장소(숫자형을 저장하는 형태)

# 문자열 : 종류 -> 문자열 저장소, 형태 -> 인덱스 사용(인덱싱/슬라이싱 가능), 불변
# --> 값을 변경할 수 없다

# 리스트 : 종류 -> 아무거나 다 저장 , 형태 -> 인덱스 사용(인덱싱/슬라이싱 가능), 가변
# --> insert / del /  append / extend / pop / remove / clear

# 튜플   : 종류 ->  아무거나 다 저장 , 형태 -> 인덱스 사용(인덱싱/슬라이싱 가능), 불변
# --> 값을 변경할 수 없다

# 딕셔너리 : 종류 -> value는 아무거나 다 저장 , 형태 -> 인덱스 없다(인덱싱/슬라이싱 없다), 가변
# --> 인덱스가 없지만, key로 접근한다.

# 집합 : 종류 -> 아무거나 다 저장 , 형태 -> 인덱스, key 없다, 중복 허용 안된다
# -> 유니크한 데이터만 저장하여 순서대로 처리함. (그래서 중복데이터가 존재하면 안됨.)

# 인덱스 : 0부터 시작하고 1씩 증가하는 값을 갖는 놈(규칙 있다)
#         저장소 접근 사용
# 인덱스 사용 예 : names = ["홍길동", "고길동", "마이꼴"]
# , kors = [80, 90, 60]
# names[0] <- "홍길동"
# kors[2] 
```

```python
# 1. 딕셔너리 생성 방법 {키1:값1, 키2:값2, ...}
# 국어점수
kors = [80, 90, 60]
print(kors[2]) 
# 2. 딕셔너리 처리
kors_dict = {
    "홍길동": 80, 
    "고길동": 90, 
    "마이꼴": 60
}
print(kors_dict["마이꼴"]) # "마이꼴" key 값으로 마이꼴의 국어 
```

```python
# 1. 딕셔너리 쌍(key:value) 추가하기. key가 명확하게 구분할 수 있는 값을 사용하는 것이 좋다
a = {1:"a"}
print(a[1])
# 추가 2:"b"
a[2] = "b"
print(a)
print(a[2])
# 추가 : name:"pey"
a["name"] = "pey"
print(a)
print(a["name"])
# 추가 리스트 : 3:[1,2,3]
a[3] = [1,2,3]
print(a)
```

```python
# 2. 딕셔너리 value 삭제
del a[1]
print(a)
```

```python
# 딕셔너리에서  key 사용해 value 얻기
# pey 점수 10 , julliet 점수 99 저장하는 딕셔너리 생성
grade = {
    'pey': 10,
    'julliet': 99
}
print(grade)
grade['julliet']
```

```python
# 딕셔너리 만들때 주의할 사항
# key 값은 중복되면 안된다
# 중복된 키 값을 가지고 있는 원래 값이 삭제된다.(덮어쓰기 한다.)
a = {1:'가'}
print(a[1])
# 추가 1 : '나'
a[1] = '나'
print(a)
b=2222
a[b] = '다'
print(a)

# key값에 리스트는 안됨. 안되는 이유는 key에 변경할 수 있는 값을 넣을수 없다.
# key값에 리스트는 들어갈 수 없음
# 튜플은 가능하다. 튜플()은 변경 불가하기 때문이다.
a[(3)] = '라' 
print(a)
```

```python
# 메소드(method) < 함수(function) len() : 메소드(식사메뉴) < 함수(식당)로 이해하면 된다.
# function : 입력(재료)받아 처리(기능)하고 반환(return)하는것이 함수다
# 함수에는 반드시 return이 존재해야 한다.
# 함수 : 함수는 데이터를 넣으면 규칙(코드)에 따라 결과물을 반환
# 함수의 구조(자판기)
# - 입력값 : 자판기에 넣는 돈과 버튼 누르기
# - 함수 내부(코드) : 자판기 안에서 음료수를 떨어뜨리는 기계 장치
# - 반환값 : 자판기 툭 떨어지는 음료수

# 함수를 왜 사용하는데? : 코드의 반복 줄이기 위해 사용
# 



```

#### 함수와 메소드 차이점 요약
|구분|함수(Funtion)|메소드(Method)|
|----|----|----|
|소속|특정 객체에 종속되지 않고 독립적으로 존재|클래스 내부에 정의되어 객체에 종속|
|호출방법|함수이름() 형태로 직접 호출|객체이름.메소드이름() 형태로 객체를 통해 호출|


```python
# 딕셔너리 메소드 : 암기
# 1. keys : 키를 반환한다 
a = {
    'name': 'pey',
    'phone': '01012341234',
    'birth' : '1118'
}
print(a)
keys_a = list(a.keys()) #keys는 리스트가 아니다
print(keys_a , type(keys_a))
print(keys_a[0])
print(keys_a[0:2]) #슬라이싱
print(a.keys()) #리스트가 아님
# print(a.values())

```

```python
# values 추출 메소드 :  values() => 값들을 리스트에 담아서 반환
# print(a.values())
# b = list(a.values())
# b[0]
# list(a.values())[0]
print(a.items())
for t_ in a.items():
    print(t_)

```

```python
# key 와 value 한쌍으로 추출 메소드 : items() 
print(a.items())
```

```python
# 모두 지우기 : clear()
print(a) #{'name': 'pey', 'phone': '01012341234', 'birth': '1118'}
a.clear() # key와 value 모두 지우기
print(a) #{} 딕셔너리는 그대로 남아있음. 
```

```python
# key로 value 얻기 :  get(key값) => return value
a = {
    'name': 'pey',
    'phone': '01012341234',
    'birth' : '1118'
}
print(a)
print(a["name"]) # 대괄호 연산자로 추출 방식
print(a.get("name")) # get() 메소드 추출 방식

# print(a['names'])
print(a.get("names")) # get() 메소드 추출 방식

if a.get("names"):
    print("이름이 있습니다.")
else:
    print("이름이 없습니다.")
```

```python
# 해당 key가 해당 딕셔너리 안에 있는지 확인 : in
print(a)
'name' in a #names라는 키가  a 딕셔너리에 키로 존재하니?
# 해당키가 있으면 True, 없으면 False 반환
# bool 타입 :  True / False -> 파이썬에선 대문자로 시작하는 True / False로 표현한다.
# print('names' in a)
if 'name' in a:
    print("name key 가 있습니다.")
else:
    print("name key 가 없습니다.")
   
```
