```python
# 딕셔너리 : 데이터 타입 -> 저장소 : 변수(속성 attribute)
# 데이터 타입별로 저장하는 방식과 종류 , 형태가 상이하다
# 숫자형 : 정수 , 실수 저장소(숫자형을 저장하는 형태)

# 문자열 : 종류 -> 문자열 저장소, 형태 -> 인덱스 사용(인덱싱/슬라이싱 가능), 불변
# --> 값을 변경할 수 없다

# 리스트 : 종류 -> 아무거나 다 저장 , 형태 -> 인덱스 사용(인덱싱/슬라이싱 가능), 가변
# --> insert / del /  append / extend / pop / remove / clear

# 튜플   : 종류 ->  아무거나 다 저장 , 형태 -> 인덱스 사용(인덱싱/슬라이싱 가능), 불변
# --> 값을 변경할 수 없다

# 딕셔너리 : 종류 -> value는 아무거나 다 저장 , 형태 -> 인덱스 없다(인덱싱/슬라이싱 없다), 가변
# --> 인덱스가 없지만, key로 접근한다.

# 집합 : 종류 -> 아무거나 다 저장 , 형태 -> 인덱스, key 없다, 중복 허용 안된다
# -> 유니크한 데이터만 저장하여 순서대로 처리함. (그래서 중복데이터가 존재하면 안됨.)

# 인덱스 : 0부터 시작하고 1씩 증가하는 값을 갖는 놈(규칙 있다)
#         저장소 접근 사용
# 인덱스 사용 예 : names = ["홍길동", "고길동", "마이꼴"]
# , kors = [80, 90, 60]
# names[0] <- "홍길동"
# kors[2] 
```

```python
# 1. 딕셔너리 생성 방법 {키1:값1, 키2:값2, ...}
# 국어점수
kors = [80, 90, 60]
print(kors[2]) 
# 2. 딕셔너리 처리
kors_dict = {
    "홍길동": 80, 
    "고길동": 90, 
    "마이꼴": 60
}
print(kors_dict["마이꼴"]) # "마이꼴" key 값으로 마이꼴의 국어 
```

```python
# 1. 딕셔너리 쌍(key:value) 추가하기. key가 명확하게 구분할 수 있는 값을 사용하는 것이 좋다
a = {1:"a"}
print(a[1])
# 추가 2:"b"
a[2] = "b"
print(a)
print(a[2])
# 추가 : name:"pey"
a["name"] = "pey"
print(a)
print(a["name"])
# 추가 리스트 : 3:[1,2,3]
a[3] = [1,2,3]
print(a)
```

```python
# 2. 딕셔너리 value 삭제
del a[1]
print(a)
```

```python
# 딕셔너리에서  key 사용해 value 얻기
# pey 점수 10 , julliet 점수 99 저장하는 딕셔너리 생성
grade = {
    'pey': 10,
    'julliet': 99
}
print(grade)
grade['julliet']
```

```python
# 딕셔너리 만들때 주의할 사항
# key 값은 중복되면 안된다
# 중복된 키 값을 가지고 있는 원래 값이 삭제된다.(덮어쓰기 한다.)
a = {1:'가'}
print(a[1])
# 추가 1 : '나'
a[1] = '나'
print(a)
b=2222
a[b] = '다'
print(a)

# key값에 리스트는 안됨. 안되는 이유는 key에 변경할 수 있는 값을 넣을수 없다.
# key값에 리스트는 들어갈 수 없음
# 튜플은 가능하다. 튜플()은 변경 불가하기 때문이다.
a[(3)] = '라' 
print(a)
```
